# `Economic Performance Analysis`

## Introduction

This report provides a comprehensive analysis of global economic performance using GDP per capita as the primary metric. We focus on the following key aspects:

1. **GDP per Capita Percentage Share for 2012:** Distribution of GDP per capita among North America, Europe, and the Rest of the World.
2. **Average GDP per Capita by Continent and Year (2004-2012):** Trends and comparisons of average GDP per capita over time.
3. **Median GDP per Capita by Continent and Year (2004-2012):** Insights into typical economic performance, mitigating the impact of outliers.

_ _ _ 

Environment Setup and Data Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure the style of the plots
sns.set(style="whitegrid")

# Load data from CSV files
df_question_3 = pd.read_csv('question_3.csv')
df_question_4 = pd.read_csv('question_4.csv')
df_question_5 = pd.read_csv('question_5.csv')
```

_ _ _

## `GDP per Capita Percentage Share (2012)`

This section analyzes the GDP per capita percentage share for North America, Europe, and the Rest of the World in 2012. Understanding the distribution of GDP per capita helps us comprehend how wealth is spread across different regions.

```python
# Preview the data for question 3
df_question_3.head()

# Plot the GDP per capita percentage share
plt.figure(figsize=(10, 6))
sns.barplot(data=df_question_3, x='percentage_share', y='continent_name', hue='continent_name', dodge=False, palette='Set2', legend=False)
plt.title('GDP per Capita Percentage Share by Region (2012)')
plt.xlabel('Percentage Share of GDP per Capita (%)')
plt.ylabel('Region')
plt.show()
```

### Description:
Europe has the highest share of GDP per capita, followed by the Rest of the World and North America. This indicates a significant disparity in economic performance across regions.

### Implications:
The high GDP per capita share in Europe suggests a concentration of wealth in this region. Stakeholders should consider the economic policies and factors contributing to this concentration and explore opportunities to foster economic growth in regions with lower GDP per capita shares.

### Recommendations:
1. **Policy Reforms:** Implement economic policies aimed at boosting GDP per capita in North America and other underperforming regions.
2. **Investment Opportunities:** Identify and invest in sectors with high growth potential in regions with lower GDP per capita.

_ _ _ 

## `Average GDP per Capita by Continent and Year (2004-2012)`

This section presents the average GDP per capita for each continent from 2004 to 2012. Tracking the average GDP per capita over the years allows us to identify economic trends and compare economic performance between continents.

```python
# Preview the data for question 4
df_question_4.head()

# Plot the average GDP per capita by continent and year
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_question_4, x='year', y='average_gdp_per_capita', hue='continent_name', marker='o')
plt.title('Average GDP per Capita by Continent (2004-2012)')
plt.xlabel('Year')
plt.ylabel('Average GDP per Capita')
plt.legend(title='Continent')
plt.grid(True)
plt.show()
```

### Description:
The average GDP per capita data reveals distinct economic growth trends for each continent. Europe consistently shows the highest average GDP per capita, highlighting its strong economic position.

### Implications:
The consistent high average GDP per capita in Europe suggests stable economic growth. Other regions show varying trends, indicating different economic challenges and opportunities.

### Recommendations:
1. **Growth Strategies:** Develop targeted economic growth strategies for continents with lower average GDP per capita.
2. **Monitoring:** Continuously monitor economic performance to identify emerging trends and adjust policies accordingly.

_ _ _ 

## `Median GDP per Capita by Continent and Year (2004-2012)`

This section analyzes the median GDP per capita for each continent from 2004 to 2012. The median provides a more accurate representation of typical economic performance by mitigating the impact of outliers.

```python
# Preview the data for question 5
df_question_5.head()

# Plot the median GDP per capita by continent and year
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_question_5, x='year', y='median_gdp_per_capita', hue='continent_name', marker='o')
plt.title('Median GDP per Capita by Continent (2004-2012)')
plt.xlabel('Year')
plt.ylabel('Median GDP per Capita')
plt.legend(title='Continent')
plt.grid(True)
plt.show()
```

### Description:
The median GDP per capita data supports the average GDP findings, providing a more nuanced view by reducing the impact of outliers. This metric confirms the economic disparity among continents.

### Implications:
The median GDP per capita is a more reliable indicator of the typical economic condition, as it reduces the influence of extreme values. This reinforces the need to address economic disparities.

### Recommendations:
1. **Inclusive Policies:** Promote inclusive economic policies that target median GDP per capita improvement.
2. **Resource Allocation:** Allocate resources effectively to regions that show significant disparities in median GDP per capita.
